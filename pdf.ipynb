```python
import re
from ase import Atoms
from PyPDF2 import PdfReader
```

```python
def get_compound_list(path) -> list:
    pattern = r'\b(?:[A-Z][a-z]*\d*)+\b'
    reader = PdfReader(path)
    pages = reader.pages
    text = ""
    for page in pages:
        sub = page.extract_text()
        text += sub
    matches = re.findall(pattern, text)

    valid_compounds = []
    for match in matches:
        try:
            mol = Atoms(match)
            valid_compounds.append(match)
        except:
            pass        

    valid_compounds = list(set(valid_compounds))
    valid_compounds = [comp for comp in valid_compounds if len(comp) > 2]

    return valid_compounds
```

```python
from pymatgen.core.composition import Composition
from pymatgen.core.periodic_table import Element
import pandas as pd

def count_elements(compound):
    element_counts = {str(el): 0 for el in Element}
    composition = Composition(compound)
    for element, count in composition.items():
        element_counts[str(element)] += count
    element_counts = pd.DataFrame.from_dict(element_counts, orient='index').T.astype(int)
    return element_counts

element_counts = {str(el): 0 for el in Element}
compound = "Cu52Ni30Fe18"
element_counts = count_elements(compound)
element_counts.value_counts()
```
